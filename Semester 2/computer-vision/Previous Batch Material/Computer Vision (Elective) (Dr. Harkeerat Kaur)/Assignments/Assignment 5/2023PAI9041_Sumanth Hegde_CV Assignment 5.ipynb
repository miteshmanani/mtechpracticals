```python
# Answer 1


import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to apply Harris Corner or SIFT
def extract_features(image, method='SIFT'):
    if method == 'Harris':
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        gray = np.float32(gray)
        dst = cv2.cornerHarris(gray, 2, 3, 0.04)
        dst = cv2.dilate(dst, None)
        keypoints = np.argwhere(dst > 0.01 * dst.max())
        keypoints = [cv2.KeyPoint(pt[1], pt[0], 1) for pt in keypoints]
        return keypoints, None
    elif method == 'SIFT':
        sift = cv2.SIFT_create()
        keypoints, descriptors = sift.detectAndCompute(image, None)
        return keypoints, descriptors
    else:
        raise ValueError('Method must be "Harris" or "SIFT"')

# Function to manually select points
def manually_select_points(image, keypoints):
    # Placeholder: select the first N points
    points = np.array([kp.pt for kp in keypoints[:10]])
    return points

# Function to find homography
def find_homography(src_points, dst_points):
    H, _ = cv2.findHomography(src_points, dst_points, cv2.RANSAC, 5.0)
    return H

# Function to stitch images
def stitch_images(image1, image2, H):
    h1, w1 = image1.shape[:2]
    h2, w2 = image2.shape[:2]

    corners_image2 = np.float32([[0, 0], [0, h2], [w2, h2], [w2, 0]]).reshape(-1, 1, 2)
    transformed_corners_image2 = cv2.perspectiveTransform(corners_image2, H)
    corners_image1 = np.float32([[0, 0], [0, h1], [w1, h1], [w1, 0]]).reshape(-1, 1, 2)

    all_corners = np.concatenate((corners_image1, transformed_corners_image2), axis=0)

    [x_min, y_min] = np.int32(all_corners.min(axis=0).ravel() - 0.5)
    [x_max, y_max] = np.int32(all_corners.max(axis=0).ravel() + 0.5)

    translation_dist = [-x_min, -y_min]

    H_translation = np.array([[1, 0, translation_dist[0]], [0, 1, translation_dist[1]], [0, 0, 1]])

    output_image = cv2.warpPerspective(image2, H_translation.dot(H), (x_max - x_min, y_max - y_min))
    output_image[translation_dist[1]:translation_dist[1] + h1, translation_dist[0]:translation_dist[0] + w1] = image1

    return output_image

# Main function for manual stitching
def manual_stitch(images):
    keypoints_list = []
    for image in images:
        keypoints, _ = extract_features(image, method='SIFT')
        keypoints_list.append(keypoints)

    points_list = []
    for keypoints in keypoints_list:
        points = manually_select_points(images[0], keypoints)
        points_list.append(points)

    H_list = []
    for i in range(len(images) - 1):
        H = find_homography(points_list[i], points_list[i + 1])
        H_list.append(H)

    result = images[0]
    for i in range(1, len(images)):
        result = stitch_images(result, images[i], H_list[i - 1])

    return result

# Main function for automatic stitching
def automatic_stitch(images):
    keypoints_list = []
    descriptors_list = []
    for image in images:
        keypoints, descriptors = extract_features(image, method='SIFT')
        keypoints_list.append(keypoints)
        descriptors_list.append(descriptors)

    bf = cv2.BFMatcher()
    matches_list = []
    for i in range(len(images) - 1):
        matches = bf.knnMatch(descriptors_list[i], descriptors_list[i + 1], k=2)
        good_matches = []
        for m, n in matches:
            if m.distance < 0.75 * n.distance:
                good_matches.append(m)
        matches_list.append(good_matches)

    H_list = []
    for i in range(len(matches_list)):
        src_pts = np.float32([keypoints_list[i][m.queryIdx].pt for m in matches_list[i]]).reshape(-1, 1, 2)
        dst_pts = np.float32([keypoints_list[i + 1][m.trainIdx].pt for m in matches_list[i]]).reshape(-1, 1, 2)
        H = find_homography(src_pts, dst_pts)
        H_list.append(H)

    result = images[0]
    for i in range(1, len(images)):
        result = stitch_images(result, images[i], H_list[i - 1])

    return result

# Load images
images = [cv2.imread('1.jpg'), cv2.imread('2.jpg'), cv2.imread('3.jpg')]

# Manual Method
manual_result = manual_stitch(images)
cv2.imwrite('manual_stitched.jpg', manual_result)

# Automatic Method
automatic_result = automatic_stitch(images)
cv2.imwrite('automatic_stitched.jpg', automatic_result)

# Display results
plt.subplot(121),plt.imshow(cv2.cvtColor(manual_result, cv2.COLOR_BGR2RGB)),plt.title('Manual Stitching')
plt.subplot(122),plt.imshow(cv2.cvtColor(automatic_result, cv2.COLOR_BGR2RGB)),plt.title('Automatic Stitching')
plt.show()
```

```python
# Answer 2






# Rectify Images and Display Epipolar Lines

import cv2
import numpy as np
import matplotlib.pyplot as plt

def extract_features_and_match(img1, img2):
    # Initialize SIFT detector
    sift = cv2.SIFT_create()

    # Find the keypoints and descriptors with SIFT
    keypoints1, descriptors1 = sift.detectAndCompute(img1, None)
    keypoints2, descriptors2 = sift.detectAndCompute(img2, None)

    # Use FLANN-based matcher for feature matching
    index_params = dict(algorithm=1, trees=5)
    search_params = dict(checks=50)
    flann = cv2.FlannBasedMatcher(index_params, search_params)

    matches = flann.knnMatch(descriptors1, descriptors2, k=2)

    # Apply ratio test
    good_matches = []
    pts1 = []
    pts2 = []

    for m, n in matches:
        if m.distance < 0.75 * n.distance:
            good_matches.append(m)
            pts1.append(keypoints1[m.queryIdx].pt)
            pts2.append(keypoints2[m.trainIdx].pt)

    pts1 = np.float32(pts1)
    pts2 = np.float32(pts2)

    return pts1, pts2, good_matches, keypoints1, keypoints2

def draw_matches(img1, img2, keypoints1, keypoints2, matches):
    match_img = cv2.drawMatches(img1, keypoints1, img2, keypoints2, matches, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
    return match_img

def rectify_images(img1, img2, pts1, pts2):
    h1, w1 = img1.shape[:2]
    h2, w2 = img2.shape[:2]

    # Find the fundamental matrix
    F, mask = cv2.findFundamentalMat(pts1, pts2, cv2.FM_RANSAC)
    if F is None:
        raise ValueError("Fundamental matrix not found. Check the input points for valid correspondences.")

    # We select only inlier points
    pts1 = pts1[mask.ravel() == 1]
    pts2 = pts2[mask.ravel() == 1]

    # Find the homography matrices for rectification
    retval, H1, H2 = cv2.stereoRectifyUncalibrated(pts1, pts2, F, imgSize=(w1, h1))

    if not retval:
        raise ValueError("Stereo rectification failed. Check the input points and images.")

    img1_rectified = cv2.warpPerspective(img1, H1, (w1, h1))
    img2_rectified = cv2.warpPerspective(img2, H2, (w2, h2))

    return img1_rectified, img2_rectified, H1, H2

def draw_epipolar_lines(img1, img2, pts1, pts2):
    F, mask = cv2.findFundamentalMat(pts1, pts2, cv2.FM_RANSAC)
    if F is None:
        raise ValueError("Fundamental matrix not found. Check the input points for valid correspondences.")

    lines1 = cv2.computeCorrespondEpilines(pts2.reshape(-1, 1, 2), 2, F)
    lines1 = lines1.reshape(-1, 3)
    img1_lines = draw_lines(img1, lines1, pts1)

    lines2 = cv2.computeCorrespondEpilines(pts1.reshape(-1, 1, 2), 1, F)
    lines2 = lines2.reshape(-1, 3)
    img2_lines = draw_lines(img2, lines2, pts2)

    return img1_lines, img2_lines

def draw_lines(img, lines, pts):
    r, c = img.shape[:2]
    img_lines = img.copy()
    for r, pt in zip(lines, pts):
        color = tuple(np.random.randint(0, 255, 3).tolist())
        x0, y0 = map(int, [0, -r[2] / r[1]])
        x1, y1 = map(int, [c, -(r[2] + r[0] * c) / r[1]])
        img_lines = cv2.line(img_lines, (x0, y0), (x1, y1), color, 1)
        img_lines = cv2.circle(img_lines, (int(pt[0]), int(pt[1])), 5, color, -1)
    return img_lines

# Load your images
img1 = cv2.imread('2.jpg')
img2 = cv2.imread('3.jpg')

# Extract features and match them
pts1, pts2, good_matches, keypoints1, keypoints2 = extract_features_and_match(img1, img2)

# Draw and display matches
match_img = draw_matches(img1, img2, keypoints1, keypoints2, good_matches)
plt.imshow(cv2.cvtColor(match_img, cv2.COLOR_BGR2RGB))
plt.title('Feature Matches')
plt.show()

# Rectify images
img1_rectified, img2_rectified, H1, H2 = rectify_images(img1, img2, pts1, pts2)

# Draw epipolar lines
img1_lines, img2_lines = draw_epipolar_lines(img1, img2, pts1, pts2)

# Display the images with epipolar lines
plt.subplot(121), plt.imshow(cv2.cvtColor(img1_lines, cv2.COLOR_BGR2RGB)), plt.title('Epipolar Lines 1')
plt.subplot(122), plt.imshow(cv2.cvtColor(img2_lines, cv2.COLOR_BGR2RGB)), plt.title('Epipolar Lines 2')
plt.show()
```

```python
# Depth Map Estimation

# To estimate a depth map, we will use OpenCV's stereo matching functions





def compute_depth_map(img1_rectified, img2_rectified):
    # Convert images to grayscale
    img1_gray = cv2.cvtColor(img1_rectified, cv2.COLOR_BGR2GRAY)
    img2_gray = cv2.cvtColor(img2_rectified, cv2.COLOR_BGR2GRAY)

    # Create stereo block matcher
    stereo = cv2.StereoBM_create(numDisparities=16*6, blockSize=15)

    # Compute disparity map
    disparity = stereo.compute(img1_gray, img2_gray).astype(np.float32)

    # Normalize the disparity map for display
    min_disp = disparity.min()
    max_disp = disparity.max()
    disparity_normalized = (disparity - min_disp) / (max_disp - min_disp)

    return disparity, disparity_normalized

# Compute the depth map
disparity, depth_map_normalized = compute_depth_map(img1_rectified, img2_rectified)

# Display the depth map
plt.imshow(depth_map_normalized, 'gray')
plt.title('Depth Map')
plt.show()
```

```python
# Generate 3D Point Cloud

# We need to reproject the depth values to 3D space using a reprojection matrix





from mpl_toolkits.mplot3d import Axes3D

def generate_point_cloud(disparity, Q):
    # Reproject image to 3D space
    points_3D = cv2.reprojectImageTo3D(disparity, Q)
    
    # Create a mask of valid points
    mask = disparity > disparity.min()
    
    # Extract valid 3D points
    output_points = points_3D[mask]
    
    # Extract the corresponding colors
    colors = cv2.cvtColor(img1_rectified, cv2.COLOR_BGR2RGB)
    output_colors = colors[mask]
    
    return output_points, output_colors

def display_point_cloud(points, colors):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Separate the points into x, y, and z coordinates
    xs = points[:, 0]
    ys = points[:, 1]
    zs = points[:, 2]
    
    # Normalize colors to [0, 1] range
    colors = colors / 255.0
    
    ax.scatter(xs, ys, zs, c=colors, s=1)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    plt.show()

# Generate the 3D point cloud
Q = np.float32([[1, 0, 0, -img1_rectified.shape[1] // 2],
                [0, -1, 0, img1_rectified.shape[0] // 2],
                [0, 0, 0, -1], 
                [0, 0, 1, 0]])
points, colors = generate_point_cloud(disparity, Q)

# Display the 3D point cloud
display_point_cloud(points, colors)
```
