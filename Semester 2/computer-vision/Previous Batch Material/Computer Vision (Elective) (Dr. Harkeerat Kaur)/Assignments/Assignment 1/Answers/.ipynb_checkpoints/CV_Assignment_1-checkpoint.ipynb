```python
#Q1

import numpy as np
import cv2
from sklearn.cluster import KMeans
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Step 2: Read an RGB image as input
def read_image(file_path):
    return cv2.imread(file_path)[:, :, ::-1]  # Read and convert to RGB

# Step 3: Implement a function to perform k-Means image quantization
def kmeans_quantization(image, k):
    pixels = image.reshape((-1, 3))
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pixels)
    quantized_colors = kmeans.cluster_centers_.astype(int)
    labels = kmeans.labels_
    quantized_image = quantized_colors[labels].reshape(image.shape)
    return quantized_image, quantized_colors

# Step 4: Convert the input RGB image to a quantized image using varying 'k'
def quantize_and_calculate_mse(image, k_values):
    original_shape = image.shape
    mse_values = []

    for k in k_values:
        quantized_image, _ = kmeans_quantization(image, k)
        mse = mean_squared_error(image.flatten(), quantized_image.flatten())
        mse_values.append(mse)

        plt.subplot(1, len(k_values), k_values.index(k) + 1)
        plt.imshow(quantized_image)
        plt.title(f'k={k}')
        plt.axis('off')

    plt.show()

    return mse_values

# Step 5: Analyze and discuss the effectiveness and limitations of k-Means image quantization

# Step 6: Full and Final Code
if __name__ == "__main__":
    # Example usage
    image_path = 'RGB image 1.jpg'
    original_image = read_image(image_path)

    # Varying 'k' values
    k_values = [1, 2, 4, 8, 16]

    # Quantize and calculate MSE
    mse_values = quantize_and_calculate_mse(original_image, k_values)

    # Display MSE values for analysis
    for k, mse in zip(k_values, mse_values):
        print(f'MSE for k={k}: {mse}')
```

```python
#Q2

import cv2
import numpy as np
from matplotlib import pyplot as plt

# Step 1: Preprocessing and Thresholding
# Load the image
image = cv2.imread('cc.jpg', cv2.IMREAD_GRAYSCALE)

# Apply binary thresholding
_, binary_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

# Step 2: Connected Components Analysis
# Find connected components
num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary_image, connectivity=8)

# Step 3: Coloring and Saving the Output
# Create a random color map
colors = np.random.randint(0, 255, size=(num_labels, 3), dtype=np.uint8)

# Create a colored image for visualization
colored_image = np.zeros((image.shape[0], image.shape[1], 3), dtype=np.uint8)
for label in range(1, num_labels):
    colored_image[labels == label] = colors[label]

# Save and display the colored image
cv2.imwrite('cc_colored.jpg', colored_image)
plt.imshow(cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# Print the number of connected components found
print(f"Number of connected components: {num_labels - 1}")
```

```python
#Q3

import cv2
import numpy as np
from skimage.metrics import peak_signal_noise_ratio as psnr
import matplotlib.pyplot as plt

# Load the original Lena image
lena_original = cv2.imread('lena_gray_512.tif', cv2.IMREAD_GRAYSCALE)

# (a) Reduce the image size to 256x256
lena_reduced = cv2.resize(lena_original, (256, 256), interpolation=cv2.INTER_AREA)
cv2.imwrite('reduced_lena_256.tif', lena_reduced)

# (b) Compare the reduced image with 'lena_gray_256.tif'
lena_gray_256 = cv2.imread('lena_gray_256.tif', cv2.IMREAD_GRAYSCALE)

# Calculate PSNR
psnr_value = psnr(lena_gray_256, lena_reduced)
print(f"PSNR between 'lena_gray_256.tif' and 'reduced_lena_256.tif': {psnr_value:.2f}")

# Calculate Mean Squared Error (MSE) difference
mse = np.mean((lena_gray_256.astype("float") - lena_reduced.astype("float")) ** 2)
mse_image = np.abs(lena_gray_256.astype("float") - lena_reduced.astype("float"))
mse_image /= mse_image.max()

# (c) Resize the computed image back to original size using different interpolation methods
methods = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC]
method_names = ['Nearest Neighbour', 'Bilinear', 'Bicubic']

plt.figure(figsize=(12, 8))
for i, method in enumerate(methods, start=1):
    resized_image = cv2.resize(lena_reduced, (512, 512), interpolation=method)
    plt.subplot(1, 3, i)
    plt.imshow(resized_image, cmap='gray')
    plt.title(method_names[i-1])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#Q4
import cv2
import matplotlib.pyplot as plt

# Load the noisy image
noisy_image = cv2.imread('Noisy image.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian blur
filtered_image = cv2.GaussianBlur(noisy_image, (5, 5), 0)
#filtered_image = cv2.medianBlur(noisy_image, 5)  # 5 is the kernel size

# Display the original noisy image and the filtered image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(noisy_image, cmap='gray')
plt.title('Noisy Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title('Filtered Image (Gaussian Blur)')
plt.axis('off')

plt.show()
```

```python
#Q6
import numpy as np

# Define the image data
image_data = np.array([
    [167, 144, 159],
    [140, 135, 154],
    [135, 148, 148]
], dtype=np.uint8)

# Get the image dimensions
height, width = image_data.shape

# Create an empty list to store the bit planes
bit_planes = []

# Loop through each bit plane (from MSB to LSB)
for i in range(8):
    # Create an empty bit plane
    bit_plane = np.zeros((height, width), dtype=np.uint8)

    # Extract the current bit from each pixel and set the corresponding bit in the bit plane
    for y in range(height):
        for x in range(width):
            # Extract the current bit using bitwise AND operation with 2^i (left shift i by 1)
            bit_value = (image_data[y, x] >> i) & 1

            # Set the corresponding bit in the bit plane
            bit_plane[y, x] = bit_value * 255

    # Append the bit plane to the list
    bit_planes.append(bit_plane)

# Print the bit planes
for i, bit_plane in enumerate(bit_planes):
    print(f"Bit plane {i+1}:")
    print(bit_plane)
```
