```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(50)
noise = np.random.normal(0, 0.03, size=50)
y = np.sin(1 + x**2) + noise
x_train, x_test = x[:30], x[30:]
y_train, y_test = y[:30], y[30:]

# Fit a polynomial curve to the training data
degree = 4
p = np.polyfit(x_train, y_train, degree)
y_test_pred = np.polyval(p, x_test)
errors = y_test_pred - y_test
total_error = np.sum(errors)
print("Total error is :", total_error)

# Plot the data points
plt.scatter(x_train, y_train, label='Training Data', color='blue')
plt.scatter(x_test, y_test, label='Test Data', color='green')

# Plot the fitted curve
x_sorted = np.sort(np.concatenate((x_train, x_test)))
plt.plot(x_sorted, np.polyval(p, x_sorted), color='red', label=f'Fitted Polynomial Curve of Degree 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Curve Fitting')
plt.legend()
plt.grid(True)
plt.show()
```
