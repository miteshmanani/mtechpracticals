#Q1. Write the "search_key" function, that searchs for a key in a binary search tree. [5 marks]

#Example:
no. of keys: 2 \\
given key: 6, 60 \\
bst: \\
30 \\
20 \\
40 \\
70 \\
60 \\
80 \\

# Output:
6 not found \\
60 found

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.value = key

def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.value < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root

def search_key(root, key):
    if root is None or root.value == key:
        return root

    if root.value < key:
        return search_key(root.right, key)

    return search_key(root.left, key)

if __name__ == "__main__":
    keys = [30, 20, 40, 70, 60, 80]
    search_keys = [6, 60]
    root = None

    for key in keys:
        root = insert(root, key)

    for key in search_keys:
        result = search_key(root, key)
        if result is not None:
            print(f"Key {key} found")
        else:
            print(f"Key {key} not found")
```

#Q3. Construct a binary search tree with the below information. The preorder traversal of a binary search tree 10, 4, 3, 5, 11, 12. [5 marks]

```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.value = key


def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.value < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root


def search_key(root, key):
    if root is None or root.value == key:
        return root
    if root.value < key:
        return search_key(root.right, key)
    return search_key(root.left, key)


def findLCA(root, n1, n2):
    while root:
        if root.value < n1 and root.value < n2:
            root = root.right
        elif root.value > n1 and root.value > n2:
            root = root.left
        else:
            break
    return root


def preorderTraversal(root):
    if root:
        print(root.value, end=' ')
        preorderTraversal(root.left)
        preorderTraversal(root.right)


def inorderTraversal(root):
    if root:
        inorderTraversal(root.left)
        print(root.value, end=' ')
        inorderTraversal(root.right)


if __name__ == "__main__":
    keys = [10, 4, 3, 5, 11, 12]  # BST keys
    root = None

    for key in keys:
        root = insert(root, key)

    print("Preorder Traversal of the BST:")
    preorderTraversal(root)
    print("\nInorder Traversal of the BST:")
    inorderTraversal(root)


```

#Q4. Write python code to check Which of the following two trees is a binary search tree. [10 marks]
image.png

```python
class Node:
    def __init__(self, key):
        self.val = key
        self.left = None
        self.right = None

def is_bst_util(node, min_val, max_val):
    if node is None:
        return True

    if node.val <= min_val or node.val >= max_val:
        return False

    return (is_bst_util(node.left, min_val, node.val) and
            is_bst_util(node.right, node.val, max_val))

def is_bst(root):
    min_val = float('-inf')
    max_val = float('inf')
    return is_bst_util(root, min_val, max_val)

# Tree 1
tree1 = Node(60)
tree1.left = Node(40)
tree1.left.left = Node(20)
tree1.left.right = Node(50)
tree1.left.right.right = Node(62)
tree1.left.right.right.left = Node(55)
tree1.right = Node(80)
tree1.right.left = Node(70)
tree1.right.right = Node(95)
tree1.right.left.right = Node(75)

# Tree 2
tree2 = Node(50)
tree2.left = Node(25)
tree2.left.left = Node(10)
tree2.left.right = Node(40)
tree2.left.right.right = Node(45)
tree2.left.right.right.left = Node(42)
tree2.right = Node(75)
tree2.right.right = Node(90)
tree2.right.left = Node(60)
tree2.right.left.right = Node(70)

# Check if each tree is a BST
is_tree1_bst = is_bst(tree1)
is_tree2_bst = is_bst(tree2)

print("Tree 1 is a BST:", is_tree1_bst)
print("Tree 2 is a BST:", is_tree2_bst)
```

#Q5. Consider the binary tree given. write a program to list each of the following: [10 marks]
1. all nodes that are leafs,
2. all pairs of nodes that are siblings
3. all ancestors of the node with value 60
4. all descendents of the node with value 25
5. the height of the tree

image.png

```python
class Node:
    def __init__(self, key):
        self.val = key
        self.left = None
        self.right = None

def find_leaf_nodes(node):
    leaf_nodes = []

    if node is None:
        return leaf_nodes

    if node.left is None and node.right is None:
        leaf_nodes.append(node.val)

    leaf_nodes.extend(find_leaf_nodes(node.left))
    leaf_nodes.extend(find_leaf_nodes(node.right))

    return leaf_nodes

def find_sibling_nodes(root, value):
    if root is None:
        return []

    if root.left and root.right:
        if root.left.val == value:
            return [root.right.val]
        elif root.right.val == value:
            return [root.left.val]
        else:
            return find_sibling_nodes(root.left, value) + find_sibling_nodes(root.right, value)
    else:
        return find_sibling_nodes(root.left, value) + find_sibling_nodes(root.right, value)

def find_ancestor_nodes(root, value):
    if root is None:
        return []

    if root.val == value:
        return []

    if root.left and root.left.val == value or root.right and root.right.val == value:
        return [root.val]

    if value < root.val:
        return [root.val] + find_ancestor_nodes(root.left, value)
    else:
        return [root.val] + find_ancestor_nodes(root.right, value)

def find_descendant_nodes(root, value):
    if root is None:
        return []

    if root.val == value:
        return find_leaf_nodes(root)

    if value < root.val:
        return find_descendant_nodes(root.left, value)
    else:
        return find_descendant_nodes(root.right, value)

def find_height(node):
    if node is None:
        return -1

    left_height = find_height(node.left)
    right_height = find_height(node.right)

    return max(left_height, right_height) + 1

root = Node(50)
root.left = Node(25)
root.left.left = Node(10)
root.left.right = Node(40)

root.left.right.left = Node(35)
root.left.right.left.right = Node(37)
root.right = Node(75)
root.right.left = Node(60)
root.right.right = Node(90)
root.right.left.right = Node(70)

print("Leaf nodes:", find_leaf_nodes(root))

print("Sibling nodes of 25:", find_sibling_nodes(root, 25))

print("Ancestor nodes of 37:", find_ancestor_nodes(root, 37))

print("Descendant nodes of 50:", find_descendant_nodes(root, 50))

print("Height of the tree:", find_height(root))
```

#Q6. Write python program to implement the "insert" and "delete" operation on a queue using linked list. [10 marks]

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def is_empty(self):
        return self.front is None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear is None:
            self.front = self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty. Cannot dequeue.")
            return None
        else:
            removed_data = self.front.data
            self.front = self.front.next
            if self.front is None:
                self.rear = None
            return removed_data

    def display(self):
        current = self.front
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

if __name__ == "__main__":
    my_queue = Queue()
    my_queue.enqueue(10)
    my_queue.enqueue(20)
    my_queue.enqueue(30)

    print("Queue elements:")
    my_queue.display()

    print("Dequeue operation result:", my_queue.dequeue())
    print("Queue elements after dequeue:")
    my_queue.display()
```

#Q7. Write a python code to explain how the elements are inserted and deleted from a doubly linked list. [5 marks]

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def delete_node(self, key):
        current = self.head
        while current:
            if current.data == key:
                if current.prev:
                    current.prev.next = current.next
                else:
                    self.head = current.next
                if current.next:
                    current.next.prev = current.prev
                return
            current = current.next
        print("Element not found in the list.")

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next
        print()

dll = DoublyLinkedList()

while True:
    print("You can perform the following Doubly Linked List Operations:, and exit:")
    print("1. Insert at End")
    print("2. Delete Node")
    print("3. Display")
    print("4. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        item = input("Enter the element to insert at end: ")
        dll.insert_at_end(item)
        print("Element inserted successfully.")
    elif choice == '2':
        key = input("Enter the element to delete: ")
        dll.delete_node(key)
        print("Element deleted successfully.")
    elif choice == '3':
        print("Doubly Linked List:")
        dll.display()
    elif choice == '4':
        print("Exiting...")
        break
    else:
        print("Invalid choice! Please enter a valid option.")
```
