```python
import cv2
import numpy as np
from scipy.ndimage import convolve
import matplotlib.pyplot as plt

image = cv2.imread(
    'C:\\Users\\mites\\OneDrive\\Desktop\\The-famous-Lena-image-often-used-as-an-example-in-image-processing.png', cv2.IMREAD_GRAYSCALE)

noise = np.random.normal(0, 25, image.shape).astype(np.uint8)
noisy_image = cv2.add(image, noise)

box_filter_mask = np.ones((5, 5), dtype=np.float32) / 25

gaussian_filter_mask = np.array([
    [1,  4,  7,  4, 1],
    [4, 16, 26, 16, 4],
    [7, 26, 41, 26, 7],
    [4, 16, 26, 16, 4],
    [1,  4,  7,  4, 1]
], dtype=np.float32)
gaussian_filter_mask /= gaussian_filter_mask.sum()

box_filtered = convolve(noisy_image, box_filter_mask, mode='reflect')

gaussian_filtered = convolve(noisy_image, gaussian_filter_mask, mode='reflect')

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(noisy_image, cmap='gray')
axes[0].set_title("Noisy Image")
axes[0].axis('off')

axes[1].imshow(box_filtered, cmap='gray')
axes[1].set_title("Box Filtered")
axes[1].axis('off')

axes[2].imshow(gaussian_filtered, cmap='gray')
axes[2].set_title("Gaussian Filtered")
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
